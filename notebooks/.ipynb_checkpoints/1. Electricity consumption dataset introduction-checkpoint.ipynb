## 2. Apply proper format to the dataset

### 2.1 Read the dataset

```python
df = pd.read_csv('../data/01_raw/historic_demand_2009_2023.csv')
df.head()
```

We take the columns we need. We will apply Time Serie Analysis to the variable `tsd`. The Transmission System Demand is equal to the national demand plus the additional generation required to meet station load, pump storage pumping and interconnector exports. Measured in MW (Megawatts).

The other variables we need are the ones related to the time. We need `settlement_date` for the day and `settlement_period` for the hour. In `settlement_period` we have values form 1 to 48, and they indicate the exact hour, where 1 is 0:00 and 48 is 23:30.

```python
data = df[['settlement_date', 'settlement_period', 'tsd']]
data.head()
```

### 2.2 Data preparation

Now we will prepare the data for the model. To do so, I will perform the following actions:
* Find out if there is any `nan` values
* Drop rows whose `settlement_period` exceeds 48, as the API states that 48 is the maximum number of samples per day
* Remove outliers


Let's begin checking if there are`nan` values:

```python
data.isna().any()
```

There isn't any `nan` value, so we go to the next step. We will drop the `settlement_period` that exceeds 48.

```python
data = data.drop(index=data[data["settlement_period"] > 48].index)
data.reset_index(drop=True, inplace=True)
```

### Bank Holidays

Bank holidays are very important in time series forecasting as they often affect the data values on those days. Therefore, I want to add a new column to my dataset to state whether each day was bank holiday or not. For that, I am using the package `holidays`.

Since the dataset covers electricity demand for England and Wales, the first step is to check that the two countries have the same bank holidays:

```python
import holidays
```

```python
# Compare England's and Wales' bank holiday
bank_holiday_england = holidays.UK(subdiv="England", years=range(2009, 2024), observed=True).items()
bank_holiday_wales = holidays.UK(subdiv="Wales", years=range(2009, 2024), observed=True).items()

print(bank_holiday_england == bank_holiday_wales)
```

The bank holidays are the same for UK and Wales. The next step is to extract the bank holidays and store them in the right format.

```python
# Create empty lists to store data
holiday_names = []
holiday_dates = []
holiday_names_observed = []
holiday_dates_observed = []

for date, name in sorted(bank_holiday_england):
    holiday_dates.append(date)
    holiday_names.append(name)
    # Pop the previous value as observed bank holidays takes place later
    if "Observed" in name:
        holiday_dates_observed.pop()
        holiday_names_observed.pop()

    holiday_names_observed.append(name)
    holiday_dates_observed.append(np.datetime64(date))
```

Once I've verified that the holidays are correctly loaded, one can compare the `holiday_dates` variable and the date in the dataset and store the boolean output in a new column: 

```python
data["is_holiday"] = data["settlement_date"].apply(lambda x: pd.to_datetime(x) in holiday_dates_observed)
```

```python
data["is_holiday"] = data["is_holiday"].astype(int)
```

```python
data.sample(7)
```

### Outlier analysis

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.histplot(data=data, x="tsd", bins=500, ax=ax, color="b");
```

I will not only delete the values equal to 0 but the entire day where there are values equal to 0. The reason being that otherwise the remaining daily values would not be representative of the entire day, which becomes more relevant for the SARIMA models as I will be using daily data instead of hourly.

```python
null_days = data.loc[data["tsd"] == 0.0, "settlement_date"].unique().tolist()

null_days_index = []

for day in null_days:
    null_days_index.append(data[data["settlement_date"] == day].index.tolist())

null_days_index = [item for sublist in null_days_index for item in sublist]

data.drop(index=null_days_index, inplace=True)
data.reset_index(drop=True, inplace=True)
```

```python
fig, ax = plt.subplots(figsize=(15, 5))

sns.histplot(data=data, x="tsd", bins=500, ax=ax, color="b");
```

### Feature creation

The first step in the feature creation is to change the date format to include the hourly values. The `settlement_period` values refer to how many samples have been taken per day. Given that there are 48 samples per day, each sample represents 30 minutes of the day.

```python
# Apply lambda function to turn settlement period into hours.
# Settlement period equal to 0 corresponds with 00:00:00 and each
# settlement period adds 30 minutes until settlement period 48, which
# corresponds with 23:30:00
data["period_hour"] = (data["settlement_period"]).apply(
    lambda x: str(timedelta(hours=(x - 1) * 0.5))
)

data.loc[data["period_hour"] == "1 day, 0:00:00", "period_hour"] = "0:00:00"

# Move the new column
column_to_move = data.pop("period_hour")
data.insert(2, "period_hour", column_to_move)

data.sample(n=7)
```

As can be seen, a new column called `period_hour` includes the hour at which the electricity demand measurement was taken. One can now combine it with the actual date as follows:

```python
data["settlement_date"] = pd.to_datetime((data["settlement_date"] + " " + data["period_hour"]))

data.sample(n=5)
```

The `settlement_date` column is now in the right format to use an index:

```python
data.set_index("settlement_date", inplace=True)
data.sort_index(inplace=True)
```

### Reducing the dataset

I will reduce the frequency of the dataset to daily summed samples. We will use this data set for the _SARIMA_ and _Exponential Smoothing_ models. The data has multiple seasonalities (daily, weekly and yearly), so we will filter it and remove what is essentially "noise" in a log time series. We will sum the values instead of the mean, because there migth be a spikes at certain hours of the day that can be masked by the average.

```python
summed_data = data["tsd"].resample("D").sum().to_frame()
summed_data["settlement_date"] = summed_data.index
summed_data["is_holiday"] = summed_data["settlement_date"].apply(lambda x: pd.to_datetime(x) in holiday_dates_observed)
summed_data["is_holiday"] = summed_data["is_holiday"].astype(int)
summed_data.drop(['settlement_date'], axis=1, inplace=True)
summed_data
```

```python
summed_data["tsd"].plot(
    style=".",
    figsize=(12, 5),
    ylabel="Total electricity demand (MW)",
    title="Daily total electricity demand",
);
```

There are values equal to 0. These values are related to the original `nan` values that were replaced before. While the data for the days containing at least one `nan` value were removed, re-sampling the data using a daily frequency leads to pandas creating those days again and assigning a value equal to 0. I will replace those values using the mean monthly value for that year.

```python
summed_data["month"] = summed_data.index.month
summed_data["year"] = summed_data.index.year
```

```python
null_info = zip(
    summed_data.loc[null_days].index.month, summed_data.loc[null_days].index.year
)

mean_values = [
    summed_data.loc[
        (summed_data["month"] == info[0]) & (summed_data["year"] == info[1]), "tsd"
    ].mean()
    for info in null_info
]

summed_data.loc[null_days, "tsd"] = mean_values

summed_data["tsd"].plot(
    style=".",
    figsize=(12, 5),
    ylabel="Total electricity demand (MW)",
    title="Daily total electricity demand",
);
```
